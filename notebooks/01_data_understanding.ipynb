```python
agrupados_por_restaurante = df.groupby("Yelp URL") #agrupo por restaurante
rating_promedio = agrupados_por_restaurante["Rating"].mean() #Agregacion para calcular el rating promedio
cantidad_reviews = agrupados_por_restaurante["Review Text"].count() #Agregacion para calcular la cantidad de reviews
#Creo un nuevo dataframe que resume las agregaciones creadas
restaurantes_info = pd.DataFrame({
    "cant_reviews": cantidad_reviews,
    "rating_promedio": rating_promedio
})
```

```python
#Me quedo con los que tengan al menos 50 reviews
restaurantes_info = restaurantes_info[restaurantes_info["cant_reviews"] >= 50]
#Ordeno los restaurantes por rating promedio de mayor a menor
restaurantes_info =restaurantes_info.sort_values("rating_promedio", ascending=False)
```
